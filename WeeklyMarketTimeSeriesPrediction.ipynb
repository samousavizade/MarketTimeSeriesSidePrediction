# LDA and QDA

## Linear Discriminant Analysis (LDA)
Linear Discriminant Analysis is a supervised machine learning algorithm used for dimensionality reduction and classification. Its primary goal is to find a linear combination of features that best separates two or more classes in the data. LDA is particularly useful when you have labeled data and want to reduce the dimensionality while preserving class separability.

## Quadratic Discriminant Analysis (QDA)

Quadratic Discriminant Analysis is another supervised classification and dimensionality reduction technique used for similar purposes as LDA. However, it differs from LDA in a few key ways.

Contrast the key differences between Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA). How do these differences impact the decision boundaries and flexibility of the two algorithms?

Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA) are both statistical (**Parametric**) models used in the field of machine learning for classification purposes. While they share some similarities, there are key differences that significantly impact their decision boundaries, flexibility, and overall application.

### Key Differences between LDA and QDA:

1. **Assumption about Covariance Matrices:**
   - **LDA:** Assumes that all classes have the same covariance matrix. This means LDA considers that the data in different classes have the same level of spread (variance) in every direction.
   - **QDA:** Does not make this assumption. Each class can have its distinct covariance matrix in QDA, implying that each class can have a different variance and covariance structure.

2. **Complexity of Decision Boundaries:**
   - **LDA:** Produces linear decision boundaries. This is because the same covariance matrix for all classes leads to a linear equation when calculating discriminant values.
   - **QDA:** Can produce quadratic (non-linear) decision boundaries. This flexibility arises from allowing different covariance matrices for each class, resulting in a quadratic equation for the discriminant function.

3. **Model Complexity and Flexibility:**
   - **LDA:** Is simpler and less flexible due to its linear nature and single covariance matrix for all classes. It is more robust to overfitting, especially in scenarios with limited training data or high-dimensional feature space.
   - **QDA:** Is more complex and flexible due to its ability to adapt to different covariance structures across classes. This flexibility allows it to fit more complex data patterns but also makes it more susceptible to overfitting, particularly in cases of limited sample size.

4. **Computational Efficiency:**
   - **LDA:** Generally more computationally efficient than QDA because it estimates fewer parameters (only one covariance matrix for all classes).
   - **QDA:** Requires estimation of a separate covariance matrix for each class, which can be computationally intensive, especially with a large number of features or classes.

### Impact on Decision Boundaries and Flexibility:

- **Decision Boundaries:**
  - LDA’s linear decision boundaries are suitable for datasets where classes are linearly separable or nearly so. However, if the actual decision boundary between classes is non-linear, LDA might not perform well.
  - QDA’s quadratic decision boundaries make it a better fit for datasets with complex class structures and non-linear relationships. It can adapt to varying shapes and orientations of the data distribution.

- **Flexibility:**
  - LDA’s lower flexibility can be an advantage in preventing overfitting and performing well in high-dimensional spaces (due to regularization effects).
  - QDA’s higher flexibility allows it to fit more complex models, but it also requires more data to estimate the parameters reliably and avoid overfitting.

If the Bayes decision boundary is linear, meaning that the optimal way to separate the classes based on the data is through a linear boundary, this has implications for the performance of both Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA) on training and test sets.

### On the Training Set:

- **LDA:** Given that LDA assumes linear decision boundaries, it aligns well with the linear nature of the Bayes decision boundary. Hence, it is expected to perform effectively on the training set, capturing the essential patterns without overfitting.
  
- **QDA:** While QDA is capable of modeling more complex (quadratic) decision boundaries, this complexity can lead to overfitting, especially in cases where the underlying true boundary is actually simpler (linear in this case). Thus, QDA might perform well on the training set but potentially capture noise and irrelevant variations.

### On the Test Set:

- **LDA:** Typically, LDA is expected to perform better on the test set when the Bayes decision boundary is linear. Its assumption about the covariance structure and linear decision boundaries generally leads to better generalization, making it robust in predicting unseen data.

- **QDA:** Given its tendency to overfit more complex models to the data, QDA might not perform as well as LDA on the test set in this scenario. The model’s complexity can result in it being too tailored to the training data, which reduces its ability to generalize to new data effectively.

If the Bayes decision boundary is non-linear, do we expect LDA or QDA to perform better on the training set? On the test set?

If the Bayes decision boundary is non-linear, the performance of Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA) will differ significantly both on the training and test sets, given their distinct assumptions about the decision boundary.

### On the Training Set:

- **LDA:** Since LDA assumes linear decision boundaries, its ability to capture the intricacies of a non-linear Bayes decision boundary is limited. As a result, it may underfit the training data, failing to model the complex relationships present.

- **QDA:** QDA, which allows for quadratic (and thereby more complex and flexible) decision boundaries, is better equipped to capture the non-linear patterns in the data. Therefore, it is expected to perform better than LDA on the training set in this scenario.

### On the Test Set:

- **LDA:** LDA's performance on the test set is likely to be suboptimal when the Bayes boundary is non-linear. Its linear model may not generalize well to new data that follows the same non-linear patterns, leading to poorer predictive performance.

- **QDA:** If the non-linear Bayes boundary can be reasonably approximated by a quadratic decision boundary, QDA is expected to perform better on the test set compared to LDA. Its ability to capture more complex patterns will be beneficial in accurately predicting outcomes on unseen data, provided that it doesn't overfit to the training data.

## b) Use the full data set to perform a logistic regression with Direction as the response and the five lag variables plus Volume as predictors. Use the summary function to print the results

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm

X = weekly[['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']]
y = weekly['Direction'].apply(lambda x: 1 if x == 'Up' else 0)

# Fitting the logistic regression model
logit_model = sm.Logit(y, sm.add_constant(X))
result = logit_model.fit()

result_summary = result.summary()
result_summary
```

The results of the logistic regression with 'Direction' as the response variable and the five lag variables plus 'Volume' as predictors are as follows:

- **Model Performance**:
  - The Pseudo R-squared value is 0.0066, indicating a very low explanatory power of the model.

- **Coefficients**:
  1. **Lag1**: Coefficient = -0.0413, p-value = 0.118. This suggests that Lag1 is not statistically significant at the 5% level.
  2. **Lag2**: Coefficient = 0.0584, p-value = 0.030. Lag2 is statistically significant, indicating a positive relationship with the probability of the market going up.
  3. **Lag3**: Coefficient = -0.0161, p-value = 0.547. This is not statistically significant.
  4. **Lag4**: Coefficient = -0.0278, p-value = 0.294. Also not statistically significant.
  5. **Lag5**: Coefficient = -0.0145, p-value = 0.583. This is not statistically significant.
  6. **Volume**: Coefficient = -0.0227, p-value = 0.538. Volume is not statistically significant in this model.

- **Intercept (Constant)**: Coefficient = 0.2669, p-value = 0.002. This is statistically significant.

The model suggests that, among all the predictors, only Lag2 has a statistically significant impact on the direction of the market (at the 5% significance level), with an increased probability of the market going up when Lag2 increases. The other lag variables and volume do not show significant predictive power for the direction of the market in this model.

Do any of the predictors appear to be statistically significant? If so, which ones?

In the logistic regression model, among all the predictors, only **Lag2** appears to be statistically significant:

- **Lag2**: It has a p-value of 0.030, which is less than the typical significance level of 0.05. This indicates that Lag2 is statistically significant. The positive coefficient (0.0584) suggests that as Lag2 increases, the probability of the market going 'Up' also increases.

The other predictors - Lag1, Lag3, Lag4, Lag5, and Volume - do not show statistical significance in this model, as their p-values are greater than the common threshold of 0.05 for significance.

## c) Compute the confusion matrix and overall fraction of correct predictions

```python
from sklearn.metrics import confusion_matrix, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)

conf_matrix, accuracy
```

The confusion matrix and the overall fraction of correct predictions are as follows:

### Confusion Matrix
- **True Negatives (Top-Left)**: 14 (actual 'Down' predicted as 'Down')
- **False Positives (Top-Right)**: 137 (actual 'Down' predicted as 'Up')
- **False Negatives (Bottom-Left)**: 19 (actual 'Up' predicted as 'Down')
- **True Positives (Bottom-Right)**: 157 (actual 'Up' predicted as 'Up')

### Overall Fraction of Correct Predictions (Accuracy)
- The model's accuracy is approximately 52.29%.

This means that the model correctly predicted the direction of the market about 52.29% of the time on the test dataset. However, it's important to note that this accuracy is only slightly better than random guessing (which would be 50% for a binary outcome), indicating that the model might not be very effective in predicting the market direction based on the given predictors.

Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

The confusion matrix provides insight into the types of mistakes made by the logistic regression model. Here's a breakdown of what each part of the matrix tells us:

2. **False Positives (137)**: These are instances where the model incorrectly predicted the market would go 'Up', but it actually went 'Down'. This is a significant number and indicates a tendency of the model to over-predict the 'Up' category.

3. **False Negatives (19)**: These are cases where the model incorrectly predicted the market would go 'Down', but it actually went 'Up'. The number here is lower compared to false positives, suggesting that the model is more cautious about predicting a 'Down' market.

4. **True Positives (157)**: These are instances where the model correctly predicted the market would go 'Up', and it did. The model performed better in predicting 'Up' movements compared to 'Down' movements.
